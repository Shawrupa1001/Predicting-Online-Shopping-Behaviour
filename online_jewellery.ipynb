# Starter Submission

## Imports

```python
import pandas as pd
import numpy as np
from sklearn.svm import SVC
```

## Paths

```python
TEST_CSV = '../input/datachallenge/online_jewellery_shop_test_final.csv'
TRAIN_CSV = '../input/datachallenge/online_jewellery_shop_train.csv'
```

## Notes

## Training

```python
train_df = pd.read_csv(TRAIN_CSV)
print(len(train_df))
train_df.head()
```

```python
print('Missing ')
train_df.loc[train_df.isnull().any(axis=1)]
```

```python
train_df = train_df.dropna(axis=0)
print(len(train_df))
print(len(train_df.columns))
```

```python
train_df_onehot = pd.get_dummies(train_df)
print(len(train_df_onehot))
print(len(train_df_onehot.columns))
train_df_onehot.head()
```

```python
X_train, y_train = train_df_onehot.drop('Revenue', axis=1), train_df_onehot['Revenue']
```

```python
classifier = SVC(probability=True)
classifier.fit(X_train, y_train)
```

```python
classifier.score(X_train, y_train)
```

## Testing

```python
test_df = pd.read_csv(TEST_CSV)
print(len(test_df))
print(len(test_df.columns))
```

```python
test_df = test_df.dropna(axis=0)
print(len(test_df))
```

```python
test_df_onehot = pd.get_dummies(test_df)
```

```python
X_test = test_df_onehot
```

```python
y_test = classifier.predict(X_test)
```

```python
y_conf = classifier.predict_proba(X_test)
y_conf_pos = y_conf[:, 1]
```

```python
df_out = pd.read_csv(TEST_CSV)[['ID']]
len(df_out)
```

```python
df_out['Revenue'] = 0
```

```python
df_out.head()
```

```python
test_df=pd.read_csv(TEST_CSV)
```

```python
non_null = (1-test_df.isnull().any(axis=1).values).astype('bool')
df_out.loc[non_null, 'Revenue'] = y_conf_pos
```
